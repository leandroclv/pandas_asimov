###iloc e filtros

```python
df = pd.DataFrame(randn(5,4), index = ['A', 'B', 'C', 'D', 'E'], columns='W X Y Z'.split())
df
```

```python
df.loc[['A', 'B'], ['W']] #df.loc[[linas], [colunas]]
```

```python
df.iloc[:-1, 1:]
```

```python
df.iloc[:-1, 1:4] #[incluso:excluso]
```

```python
df > 0
```

```python
df[df> 0]
```

```python
df['Y'] > 0
```
